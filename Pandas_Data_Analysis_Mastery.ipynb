#Filtering - Using Conditionals to Filter Rows and Columns

```python
df_people
```

```python
df_people['last'] == 'Doe'
```

```python
filt = (df_people['last'] == 'Doe')
df_people[filt]
```

#using filt as filter filter is a keyword in python


```python
df_people[df_people['last'] == 'Doe']
```

```python
df_people.loc[filt,'email']
```

```python
filt = (df_people['first'] == 'John') & (df_people['last'] == 'Doe')
df_people.loc[filt]
```

```python
filt = (df_people['first'] == 'John') | (df_people['last'] == 'Schafer')
df_people.loc[filt]
```

```python
df_people.loc[~filt]
```
